```python
import math
import numpy as np
from matplotlib import pyplot as plt 
#distances in millions of miles. 
x = [35.98,67.24,92.96,141.6,483.8,890.8,1784.0,2793.0]
# masses
merc = 0.330 * math.pow(10,24)
venus = 4.87 * math.pow(10,24)
earth = 5.97 * math.pow(10,24)
mars = 0.642 * math.pow(10,24)
jupiter = 1898 * math.pow(10,24)
saturn = 568 * math.pow(10,24)
uranus = 86.8 * math.pow(10,24)
neptune = 102 * math.pow(10,24)
y = [merc, venus, earth, mars, jupiter, saturn, uranus, neptune ]
plt.plot(x,y)
plt.title(" Distance from the Sun vs Mass of Planet")
plt.xlabel("distance from Sun (millions of mi)")
plt.ylabel("mass of planet in kg * (10^24)")
plt.show()
```
